```python
sensor_values = [2, 5, 8, 9, 12, 15]

print(sensor_values)
```

```python
threshold = 8

print("Threshold value is:", threshold)
```

```python
def simulated_fpga_threshold(value):
    if value > threshold:
        return "HIGH"
    else:
        return "NORMAL"
```

```python
print(simulated_fpga_threshold(5))
print(simulated_fpga_threshold(10))
```

```python
for value in sensor_values:
    fpga_output = simulated_fpga_threshold(value)
    print("Sensor Value:", value, "→ Simulated FPGA Output:", fpga_output)
```

```python
def ai_decision(value, fpga_output):
    if fpga_output == "NORMAL":
        return "Normal", "The sensor value is within the safe range."
    elif value <= 12:
        return "Warning", "The sensor value is above the threshold and should be monitored."
    else:
        return "Critical", "The sensor value is much higher than the threshold and needs attention."
```

```python
for value in sensor_values:
    fpga_output = simulated_fpga_threshold(value)
    decision, explanation = ai_decision(value, fpga_output)

    print("Sensor Value:", value)
    print("Simulated FPGA Output:", fpga_output)
    print("AI Decision:", decision)
    print("Explanation:", explanation)
    print("--------------------------------")
```

```python
for value in sensor_values:
    fpga_output = simulated_fpga_threshold(value)

    prompt = f"""
    A simulated FPGA system detected a sensor value of {value}.
    The threshold is {threshold}.
    The FPGA classified this value as {fpga_output}.

    Classify this as Normal, Warning, or Critical and explain briefly.
    """

    print("------ AI PROMPT ------")
    print(prompt)
    print("-----------------------\n")
```

```python
print("FINAL RESULTS")
print("Value | FPGA Output | AI Decision")
print("----------------------------------")

results = [
    (2, "NORMAL", "Normal"),
    (5, "NORMAL", "Normal"),
    (8, "NORMAL", "Normal"),
    (9, "HIGH", "Warning"),
    (12, "HIGH", "Warning"),
    (15, "HIGH", "Critical")
]

for value, fpga, ai in results:
    print(f"{value:5} | {fpga:12} | {ai}")
```
